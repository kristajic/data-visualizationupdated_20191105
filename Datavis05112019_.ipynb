**The Pandas library is one of the leading open source and free libraries to read and analyze our data from csv and similar files.**

_Name is derived from the combination of the words panel data. _

* Works fast
* Allows downloading data from different files into memory
* Provides the opportunity to work with missing data
* Enables reshaping data sets
* Can divide data into different indexes and sections based on labels
* Columns can be added and deleted
* Works with temporal data

**There are 3 basic data structures**
* 'Series': High-value Lisets containing the same type of data.
* 'Data Frame': 2D lists
* 'Panel': 3D lists

# Scatter Plot

* _Is it possible to estimate the price per square meter based on our data ?_


```python
import numpy as np
import matplotlib.pyplot as plt # we will use matplot library and you might see some examples as below.
#plt.rcParams['figure.figsize'] = [30, 30] ; you might try with [40,40] for example.

fig, ax = plt.subplots()

ax.scatter(house['LotArea'], house['SalePrice'], alpha=0.5)


ax.set(xlabel='area', ylabel='price',
       title='Price/Area')

plt.show()
```

* It is also possible to produce these graphics with pandas.

```python
ax2 = house.plot.scatter(x='LotArea',y='SalePrice')
plt.show()
```

```python
ax2 = evler.plot.scatter(x='LotArea', y='SalePrice', c='NeighborhoodID',colormap='plasma') # please try the colormaps with inferno,viridis etc.
plt.show()
```

* In order to better understand our image above, we may add something to the parameters to increase the point size. s = 500

* If we change the transparency of the dots, we can see the underlying data. alpha = 0.2

```python
house2 = house[house.NeighborhoodID ==9]
ax2 = house2.plot.scatter(x='LotArea', y='SalePrice', c='NeighborhoodID',colormap='plasma')
plt.show()
```
